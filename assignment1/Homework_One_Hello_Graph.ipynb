The code below creates a graph. I created this graph manually, to reflect my family tree - my parents, their siblings, and those siblings' children.

First, I initialize a graph.

```python
test_graph=net.Graph()
```

Then I add my nodes and edges. This is admittedly a little inconsistent, but I've created an edge for each of my parents to their siblings, and then those siblings and my parents to their children. 

```python
test_graph.add_edge('Ch','Km')
test_graph.add_edge('Ch','T')
test_graph.add_edge('Km','E')
test_graph.add_edge('E','S')
test_graph.add_edge('E','A')
test_graph.add_edge('E','N')
test_graph.add_edge('A','N')
test_graph.add_edge('S','N')
test_graph.add_edge('T','B')
test_graph.add_edge('T','G')
test_graph.add_edge('B','G')
test_graph.add_edge('Km','C')
test_graph.add_edge('Ch','C')
```

Then I can draw a simple graph of these nodes and edges.

```python
net.draw_networkx(test_graph)
```
